```python
from langchain.prompts import ChatPromptTemplate
from langchain.output_parsers import StrOutputParser
from langchain.chat_models import ChatOpenAI  # or any other model

# Step 1: Create a prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("user", "{input}")
])

# Step 2: Initialize the model
llm = ChatOpenAI()

# Step 3: Initialize the parser
parser = StrOutputParser()

# Step 4: Chain the components
chain = prompt | llm | parser

# Step 5: Run the chain
response = chain.invoke({"input": "What's the capital of Japan?"})
print(response)  # Output: "Tokyo"
```
